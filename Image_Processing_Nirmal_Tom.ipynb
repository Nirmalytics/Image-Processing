## Image Rotation


```python
# import required libraries
from skimage.transform import rotate


print('Rotated Image')
# rotating the image by 180 degrees
rotated = rotate(image,180,resize=True)

# plot the rotated image
imshow(rotated);
```

### Image Cropping


```python

# importing plotting library
import matplotlib.pyplot as plt
%matplotlib inline

# selecting part of the image only 
cropped1 = image[1000:1500,200:600]
cropped2 = image[200:900,700:1250]

# plot original and cropped images
imshow(image)
plt.title('Original Image')

# matplotlib code to display
fig,ax = plt.subplots(nrows=1,ncols=2,figsize=(10,10))
ax[0].imshow(cropped1)
ax[0].set_title('Cropped Image1')
ax[1].imshow(cropped2)
ax[1].set_title('Cropped Image2')
```

### Flipping


```python

# flip image left-to-right
flipLR = np.fliplr(image)

plt.imshow(flipLR)
plt.title('Left to Right Flipped')
```

```python
# flip image up-to-down
flipUD = np.flipud(image)

plt.imshow(flipUD)
plt.title('Up Down Flipped')
```

### Brightness Manipulation


```python
# import required libraries
from skimage.exposure import adjust_gamma

# brighten the image
bright = adjust_gamma(image,gamma=0.5,gain=1)

imshow(bright)
plt.title('Brightened IMage')
```

```python
# darken the image
dark = adjust_gamma(image,gamma=1.5,gain=1)

imshow(dark)
plt.title('Darkened Image')
```
